```python
barao = barao.drop(["Cód. Eleição", "Desc. Eleição", "Data e hora", "CNPJ Prestador Conta", "UF", "Sigla da UE",
                   "Nome da UE", "CPF do vice/suplente", "Tipo de documento", "Número do documento",
                   "Cod setor econômico do fornecedor"],
                  axis = 1)

barao
```

#### Convertendo coluna "Valor receita" e "Valor despesa" em float

```python
# Primeiro precisamos trocar a vírgula por ponto porque é o padrão no pandas
barao_rec["Valor receita"] = barao_rec["Valor receita"].apply(lambda x: str(x).replace(",", "."))

# Agora podemos converter o objeto em float
barao_rec["Valor receita"] = barao_rec["Valor receita"].astype("float64")
```

```python
# Primeiro precisamos trocar a vírgula por ponto porque é o padrão no pandas
barao["Valor despesa"] = barao["Valor despesa"].apply(lambda x: str(x).replace(",", "."))

# Agora podemos converter o objeto em float
barao["Valor despesa"] = barao["Valor despesa"].astype("float64")
```

#### Base Despesas

```python
# Separando prefeito e vereador

barao_pref = barao.loc[barao["Cargo"] == "Prefeito"]
barao_ver = barao.loc[barao["Cargo"] == "Vereador"]
```

```python
# Quem gastou mais em Barão de Cocais em 2016? Prefeitos

barao_pref["despesas"] = barao_pref.groupby(["CPF do candidato", "Sigla  Partido",
                                             "Nome candidato"])["Valor despesa"].transform("sum")
despesas_barao_pref = (barao_pref[["CPF do candidato", "Sigla  Partido", "Nome candidato",
                                  "despesas"]].drop_duplicates().sort_values("despesas", ascending = False))
despesas_barao_pref
```

```python
# Quem gastou mais em Barão de Cocais em 2016? Vereador

barao_ver["despesas"] = barao_ver.groupby(["CPF do candidato", "Sigla  Partido",
                                             "Nome candidato"])["Valor despesa"].transform("sum")
despesas_barao_ver = (barao_ver[["CPF do candidato", "Sigla  Partido", "Nome candidato",
                                  "despesas"]].drop_duplicates().sort_values("despesas", ascending = False))
despesas_barao_ver.head(50)
```

```python
# Qual a média dos gastos? - Prefeitos

despesas_barao_pref.describe()
```

```python
# Qual a média dos gastos? - Vereador

despesas_barao_ver.describe()
```

```python
# Ranking de despesas - Prefeitos

gastos_pref = barao_pref[["Tipo despesa", "Valor despesa"]]
gastos_pref = (gastos_pref.groupby(["Tipo despesa"]).sum().sort_values("Valor despesa", ascending = False))
gastos_pref
```

```python
# Ranking de despesas - Vereador

gastos_ver = barao_ver[["Tipo despesa", "Valor despesa"]]
gastos_ver = (gastos_ver.groupby(["Tipo despesa"]).sum().sort_values("Valor despesa", ascending = False))
gastos_ver
```

```python
# Quais fornecedores ganharam mais dinheiro?

barao["despesas"] = barao.groupby(["CPF/CNPJ do fornecedor", "Nome do fornecedor"])["Valor despesa"].transform("sum")
desp_barao = (barao[["CPF/CNPJ do fornecedor", "Nome do fornecedor", "despesas"]]).drop_duplicates().sort_values("despesas", ascending = False)
desp_barao.head(50)
```

```python
# Quais fornecedores ganharam mais dinheiro? - Geral

barao["despesas"] = barao.groupby(["CPF/CNPJ do fornecedor", "Nome do fornecedor","Sigla  Partido",
                                   "Nome candidato"])["Valor despesa"].transform("sum")
despesas_barao = (barao[["CPF/CNPJ do fornecedor", "Nome do fornecedor","Sigla  Partido", "Nome candidato",
                         "despesas"]]).drop_duplicates().sort_values("despesas", ascending = False)
despesas_barao.head(50)
```

```python
# Quais fornecedores ganharam mais dinheiro? - Geral

barao_pref["despesas"] = barao_pref.groupby(["CPF/CNPJ do fornecedor", "Nome do fornecedor","Sigla  Partido",
                                   "Nome candidato"])["Valor despesa"].transform("sum")
despesas_barao_pref = (barao_pref[["CPF/CNPJ do fornecedor", "Nome do fornecedor","Sigla  Partido", "Nome candidato",
                         "despesas"]]).drop_duplicates().sort_values("despesas", ascending = False)
despesas_barao_pref.head(50)
```

```python
# Quais fornecedores ganharam mais dinheiro? - Geral

barao_ver["despesas"] = barao_ver.groupby(["CPF/CNPJ do fornecedor", "Nome do fornecedor","Sigla  Partido",
                                   "Nome candidato"])["Valor despesa"].transform("sum")
despesas_barao_ver = (barao_ver[["CPF/CNPJ do fornecedor", "Nome do fornecedor","Sigla  Partido", "Nome candidato",
                         "despesas"]]).drop_duplicates().sort_values("despesas", ascending = False)
despesas_barao_ver.head(50)
```

```python
# Os candidatos gastaram com os mesmos fornecedores? - Precisa de análise
```

```python
# Média de gastos dos partidos por candidato

# 1º Gastos por partido
barao["gasto_part"] = barao.groupby(["Sigla  Partido"])["Valor despesa"].transform("sum")
gasto_partido = (barao)[["Sigla  Partido", "gasto_part"]].drop_duplicates().sort_values("gasto_part", ascending = False)

gasto_partido
```

```python
# Encontrar a quantidade de candidatos por partido
barao["candidatos"] = barao.groupby(["Sigla  Partido"])["Nome candidato"].transform("nunique")


gasto_partido = barao[["Sigla  Partido", "gasto_part", "candidatos"]].drop_duplicates()


# Calculando a média de gastos
gasto_partido["media"] = gasto_partido["gasto_part"] / gasto_partido["candidatos"]
gasto_partido = gasto_partido.sort_values(by = "media", ascending = False)
gasto_partido
```
