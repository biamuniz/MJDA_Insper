```python
import pandas as pd
import csv
```

## Bases de 2012

```python
cand12 = pd.read_csv(r"consulta_cand_2012_MG.csv", sep=";", encoding="latin_1", error_bad_lines=False)
```

```python
desp12 = pd.read_csv(r"despesas_candidatos_2012_MG.txt", sep=";", encoding="latin_1", error_bad_lines=False)
```

```python
cand12.columns #olhando as colunas do dataframe de candidatos de 2012
```

```python
desp12.columns #olhando as colunas do dataframe de candidatos de 2012
```

```python
# As planilhas possuem colunas com nomes diferentes! Precisa padronizar pelo menos aquelas colunas que serão analisadas
desp12.rename(columns={'Sequencial Candidato': 'SQ_CANDIDATO'}, inplace = True)
desp12.rename(columns={'Município': 'NM_UE'}, inplace = True)
desp12.rename(columns={'Sigla  Partido': 'SG_PARTIDO'}, inplace = True)
desp12.rename(columns={'Número candidato': 'NR_CANDIDATO'}, inplace = True)
desp12.rename(columns={'Nome candidato': 'NM_CANDIDATO'}, inplace = True)
desp12.rename(columns={'CPF do candidato': 'NR_CPF_CANDIDATO'}, inplace = True)
desp12.rename(columns={'Cargo': 'DS_CARGO'}, inplace = True)
```

```python
# Unindo a base com informações de candidatos a base de despesas. Escolhi a coluna SQ_CANDIDATO pra fazer isso
base_2012 = pd.merge(desp12, cand12, on=['SQ_CANDIDATO', 'SQ_CANDIDATO'], how='left')
```

```python
## Filtrei a base das eleições por município
itabira = base_2012.loc[base_2012["SG_UE"] == 46337]
sgrabaixo = base_2012.loc[base_2012["SG_UE"] == 52370]

# Juntei as bases das eleições nos municípios de interesse em uma só
barragem_2012 = itabira.append(sgrabaixo)

barragem_2012
```

```python
barragem_2012.columns
```

```python
# Primeiro precisamos trocar a vírgula por ponto porque é o padrão no pandas
barragem_2012["Valor despesa"] = barragem_2012["Valor despesa"].apply(lambda x: str(x).replace(",", "."))

# Agora podemos converter o objeto em float
barragem_2012["Valor despesa"] = barragem_2012["Valor despesa"].astype("float64")
```

```python
barragem_2012
```

```python
# Separando prefeito e vereador

barragem_2012_pref = barragem_2012.loc[barragem_2012["DS_CARGO_x"] == "Prefeito"]
barragem_2012_ver = barragem_2012.loc[barragem_2012["DS_CARGO_x"] == "Vereador"]
```
