**2) Filtros**: Não consegui filtrar por mais de uma condição (exemplo: filtrar todos os códigos de todos os municípios de interesse de uma vez). Então contornei com o código abaixo:

```python
# Juntei as bases das três eleições em uma só
candidatos = cand12.append(cand16)
candidatos = candidatos.append(cand20)
```

```python
## Filtrei a base das eleições por município
bcocais= candidatos.loc[candidatos["SG_UE"] == 41076]
brumadinho = candidatos.loc[candidatos["NM_UE"] == 41793]
caltas = candidatos.loc[candidatos["SG_UE"] == 41840]
itabira = candidatos.loc[candidatos["SG_UE"] == 46337]
nlima = candidatos.loc[candidatos["SG_UE"] == 48950]
mariana = candidatos.loc[candidatos["SG_UE"] == 47996]
ouropreto = candidatos.loc[candidatos["SG_UE"] == 49212]
sgrabaixo = candidatos.loc[candidatos["SG_UE"] == 52370]
```

```python
# Juntei as bases das eleições nos municípios de interesse em uma só
barragem = bcocais.append(brumadinho)
barragem = barragem.append(caltas)
barragem = barragem.append(itabira)
barragem = barragem.append(nlima)
barragem = barragem.append(mariana)
barragem = barragem.append(ouropreto)
barragem = barragem.append(sgrabaixo)
```

```python
barragem
```

**3) Exportando o arquivo final:** Um csv com os candidatos das eleições municipais em 2012, 2016 e 2020 nos municípios com barragens em MG

```python
barragem.to_csv("candidatos_barragens.csv", sep=';', encoding='utf-8')
```
