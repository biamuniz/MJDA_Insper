**1) Importando bibliotecas e arquivos**

```python
import pandas as pd
import csv
```

```python
cand12 = pd.read_csv(r"consulta_cand_2012_MG.csv", sep=";", encoding="latin_1", error_bad_lines=False)
cand16 = pd.read_csv("consulta_cand_2016_MG.csv", sep=";", encoding="latin_1", error_bad_lines=False)
cand20 = pd.read_csv(r"consulta_cand_2020_MG.csv", sep=";", encoding="latin_1", error_bad_lines=False)
desp12 = pd.read_csv(r"despesas_candidatos_2012_MG.txt", sep=";", encoding="latin_1", error_bad_lines=False)
desp16 = pd.read_csv(r"despesas_candidatos_2016_MG.txt", sep=";", encoding="latin_1", error_bad_lines=False)
desp20 = pd.read_csv(r"despesas_contratadas_candidatos_2020_MG.csv", sep=";", encoding="latin_1", error_bad_lines=False)
```

```python
desp12.rename(columns={'CPF/CNPJ do fornecedor': 'NR_CPF_CNPJ_FORNECEDOR'}, inplace = True)
desp12.rename(columns={'Município': 'NM_UE'}, inplace = True)
desp12.rename(columns={'Sigla  Partido': 'SG_PARTIDO'}, inplace = True)
desp12.rename(columns={'Número candidato': 'NR_CANDIDATO'}, inplace = True)
desp12.rename(columns={'Nome candidato': 'NM_CANDIDATO'}, inplace = True)
desp12.rename(columns={'CPF do candidato': 'NR_CPF_CANDIDATO'}, inplace = True)
desp12.rename(columns={'Cargo': 'DS_CARGO'}, inplace = True)
desp12.rename(columns={'Nome do fornecedor (Receita Federal)': 'NM_FORNECEDOR_RFB'}, inplace = True)
desp12.rename(columns={'Valor despesa': 'VR_DESPESA_CONTRATADA'}, inplace = True)
desp12.rename(columns={'Descriçao da despesa': 'DS_DESPESA'}, inplace = True)
```

```python
desp16.rename(columns={'CPF/CNPJ do fornecedor': 'NR_CPF_CNPJ_FORNECEDOR'}, inplace = True)
desp16.rename(columns={'Município': 'NM_UE'}, inplace = True)
desp16.rename(columns={'Sigla  Partido': 'SG_PARTIDO'}, inplace = True)
desp16.rename(columns={'Número candidato': 'NR_CANDIDATO'}, inplace = True)
desp16.rename(columns={'Nome candidato': 'NM_CANDIDATO'}, inplace = True)
desp16.rename(columns={'CPF do candidato': 'NR_CPF_CANDIDATO'}, inplace = True)
desp16.rename(columns={'Cargo': 'DS_CARGO'}, inplace = True)
desp16.rename(columns={'Nome do fornecedor (Receita Federal)': 'NM_FORNECEDOR_RFB'}, inplace = True)
desp16.rename(columns={'Valor despesa': 'VR_DESPESA_CONTRATADA'}, inplace = True)
desp16.rename(columns={'Descriçao da despesa': 'DS_DESPESA'}, inplace = True)
```

**2) Filtros**: Não consegui filtrar por mais de uma condição (exemplo: filtrar todos os códigos de todos os municípios de interesse de uma vez). Então contornei com o código abaixo:

```python
# Juntei as bases das três eleições em uma só
candidatos = cand12.append(cand16)
candidatos = candidatos.append(cand20)

## Filtrei a base das eleições por município
itabira = candidatos.loc[candidatos["SG_UE"] == 46337]
sgrabaixo = candidatos.loc[candidatos["SG_UE"] == 52370]

# Juntei as bases das eleições nos municípios de interesse em uma só
cand_barragem = itabira.append(sgrabaixo)
```

```python
cand_barragem
```

**3) Exportando o arquivo final:** Um csv com os candidatos das eleições municipais em 2012, 2016 e 2020 nos municípios com barragens em MG

```python
cand_barragem.to_csv("candidatos_barragens.csv", sep=';', encoding='utf-8')
```
