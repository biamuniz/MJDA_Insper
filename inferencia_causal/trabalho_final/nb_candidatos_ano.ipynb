**1) Importando bibliotecas e arquivos**

```python
import pandas as pd
import csv
```

```python
cand12 = pd.read_csv(r"consulta_cand_2012_MG.csv", sep=";", encoding="latin_1", error_bad_lines=False)
desp12 = pd.read_csv(r"despesas_candidatos_2012_MG.txt", sep=";", encoding="latin_1", error_bad_lines=False)
```

```python
desp12.rename(columns={'Sequencial Candidato': 'SQ_CANDIDATO'}, inplace = True)
desp12.rename(columns={'Município': 'NM_UE'}, inplace = True)
desp12.rename(columns={'Sigla  Partido': 'SG_PARTIDO'}, inplace = True)
desp12.rename(columns={'Número candidato': 'NR_CANDIDATO'}, inplace = True)
desp12.rename(columns={'Nome candidato': 'NM_CANDIDATO'}, inplace = True)
desp12.rename(columns={'CPF do candidato': 'NR_CPF_CANDIDATO'}, inplace = True)
desp12.rename(columns={'Cargo': 'DS_CARGO'}, inplace = True)
```

```python
base_2012 = pd.merge(desp12, cand12, on=['SQ_CANDIDATO', 'SQ_CANDIDATO'], how='left')
```

```python
## Filtrei a base das eleições por município
itabira = base_2012.loc[base_2012["SG_UE"] == 46337]
sgrabaixo = base_2012.loc[base_2012["SG_UE"] == 52370]

# Juntei as bases das eleições nos municípios de interesse em uma só
barragem_2012 = itabira.append(sgrabaixo)

barragem_2012
```
