Qual o valor total, máximo, mínimo, média e mediana de despesas por município?

```python
despesas_rj = pd.read_csv("https://raw.githubusercontent.com/biamuniz/mjda_insper/main/inferencia_causal/atividade%201/despesas_rio.csv", sep=";")
despesas_rj
```

```python
# Para responder à pergunta, fizemos uma tabela dinâmica
variaveis = despesas_rj.columns.difference(set(['muni']).union(set([])).union(set({'valor'}))) # Acho que selecionei as colunas que eu quero olhar (município e valor da despesa)
variaveis2 = despesas_rj.drop(variaveis, axis=1) # Não sei o que significa
# Fazendo a tabela dinâmica
pivot_table = variaveis2.pivot_table(
    index=['muni'],
    values=['valor'],
    aggfunc={'valor': ['sum', 'mean', 'min', 'max', 'median', 'count']}
)


# Reset the column name and the indexes
tabela_municipio = pivot_table.reset_index()
```

```python
tabela_municipio
```

```python
# Despesas por partido
unused_columns = despesas_rj.columns.difference(set(['partido']).union(set([])).union(set({'valor'})))
tmp_df = despesas_rj.drop(unused_columns, axis=1)
pivot_table = tmp_df.pivot_table(
    index=['partido'],
    values=['valor'],
    aggfunc={'valor': ['sum', 'mean', 'min', 'max', 'median', 'count']}
)


# Reset the column name and the indexes
despesa_partido = pivot_table.reset_index()
```
