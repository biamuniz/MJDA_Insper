```python
# Criando um novo dataframe ("prefeitos_rj") que vai abrigar todos os candidatos com a "sigla_uf" do Rio de Janeiro
prefeitos_rj = prefeitos.loc[prefeitos["sigla_uf"] == "RJ"]
prefeitos_rj # Para visualizar o novo dataframe
```
